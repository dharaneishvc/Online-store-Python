```python
from PyPDF2 import PdfReader, PdfWriter

def add_bookmarks(pdf_path, output_path, bookmarks):
    with open(pdf_path, 'rb') as file:
        pdf_reader = PdfReader(file)
        pdf_writer = PdfWriter()

        # Add all pages to the output PDF
        for page_number in range(len(pdf_reader.pages)):
            page = pdf_reader.pages[page_number]
            pdf_writer.add_page(page)

        # Add bookmarks for each title
        for title, page_number in bookmarks.items():
            # Create a bookmark for the title
            pdf_writer.add_outline_item(title, page_number - 1)

        with open(output_path, 'wb') as output_file:
            pdf_writer.write(output_file)

if __name__ == "__main__":
    pdf_path = "D:\Btech\Important Docs\BTech CSE 2019 Curriculum and Syllabus.pdf"  # Replace with the path to your input PDF file
    output_path = "D:\Btech\Important Docs\BTech_CSE-2019_Curriculum_Syllabus.pdf"  # Replace with the desired output path
    bookmarks = {
        'General Information': 2,
        'Curriculum': 4,
        'Semester 1': 15,
        'Semester 2': 31,
        'Semester 3': 50,
        'Semester 4': 66,
        'Semester 5': 80,
        'Semester 6': 95,
        'Semester 7': 107,
        'Semester 8': 111,
        'Electives in Cyber Security' :112,
        'Electives in Networks' :128,
        'Electives in Data Science' :136,
        'Electives in Computer Vision' :150,
        'Electives in Cyber-Physical Systems' :170,
        'Electives in Artificial Intelligence' :182,
        'Electives General' :198,
        'Free Electives':206,
        'Professional Electives for Other Branches':216
    }

    add_bookmarks(pdf_path, output_path, bookmarks)
```

```python
from PyPDF2 import PdfReader, PdfWriter

def add_nested_bookmarks(pdf_path, output_path, bookmarks):
    with open(pdf_path, 'rb') as file:
        pdf_reader = PdfReader(file)
        pdf_writer = PdfWriter()

        # Add all pages to the output PDF
        for page_number in range(len(pdf_reader.pages)):
            page = pdf_reader.pages[page_number]
            pdf_writer.add_page(page)

        # Add bookmarks for each title
        for title, (page_number, sub_bookmarks) in bookmarks.items():
            # Create a bookmark for the title
            title_page = pdf_reader.pages[page_number - 1]
            ref = pdf_writer.add_outline_item(title, page_number)

            # Add nested bookmarks for the unit
            for sub_title, sub_page_number in sub_bookmarks.items():
                sub_page = pdf_reader.pages[sub_page_number - 1]
                pdf_writer.add_outline_item(sub_title, sub_page_number, parent=ref)

        with open(output_path, 'wb') as output_file:
            pdf_writer.write(output_file)

if __name__ == "__main__":
    pdf_path = '/content/19HUM239_Psychology_full_merged.pdf'
    output_path = 'output_with_nested_bookmarks.pdf'
    bookmarks = {
        'Unit 1': (1, {
            'Abraham Maslow\'s hierarchy of motivation and needs': 1,
            'Self Esteem - importance and enhancement': 20,
            'Self analysis through SWOT': 35,
            'Johari Window': 52,
            'Self Awareness & Self Motivation': 60
        }),
        'Unit 2': (63, {
            'Stress - stressors, eustress & distress': 63,
            'Stress Management' :76,
            'PTSD': 94,
            'Guided Imaginary - Coping of Stress': 101,
            'Conflict & people': 109,
            'Empathy': 136,
            'Healthy Relationships': 151
        }),
        'Unit 3': (168, {
            'Health compromising behaviours': 168,
            'Alcohol': 176,
            'Teen Smoking': 191,
            'Substance abuse and addiction': 204
        }),
        # Add more units as needed
    }

    add_nested_bookmarks(pdf_path, output_path, bookmarks)
```
